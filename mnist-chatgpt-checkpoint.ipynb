## ChatGPT Prompt

Can you generate the code so that it saves the results after each epoch on the disk, then reads the results of the provious epoch from the disk and resumes the traning 

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint
import pandas as pd
import os
```

```python
# Solution provided 
from tensorflow.keras.datasets import mnist
from tensorflow.keras import layers, models
import os

# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Preprocess the dataset
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# Define the model
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Define the checkpoint directory and file names
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt_{epoch}")

# Create the checkpoint callback
checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True)

# Define the history file name
history_file = os.path.join(checkpoint_dir, 'training_history.csv')

# Define the CSV logger callback
csv_logger = tf.keras.callbacks.CSVLogger(history_file)

# Train the model with checkpoint and CSV logger callbacks
history = model.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels),
                    callbacks=[checkpoint_callback, csv_logger])

# Load the model weights and training history from the last checkpoint
if os.path.exists(checkpoint_dir):
    latest = tf.train.latest_checkpoint(checkpoint_dir)
    if latest is not None:
        model.load_weights(latest)
        print(f"Resumed training from checkpoint {latest}")
        history = pd.read_csv(history_file)
else:
    os.makedirs(checkpoint_dir)

# Continue training from the last epoch of the previous run
model.fit(train_images, train_labels, epochs=10, initial_epoch=len(history),
          validation_data=(test_images, test_labels),
          callbacks=[checkpoint_callback, csv_logger])
```
