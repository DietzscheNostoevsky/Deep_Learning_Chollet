Sure, I can show you a graph of the PDF and CDF of X. Here is a graph of the PDF of X, which shows the probability density of X at different values of x:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = np.linspace(-5, 5, 100)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('PDF of X')
plt.show()
```

Here is a graph of the CDF of X, which shows the cumulative probability of X being less than or equal to different values of x:

```python
x = np.linspace(-5, 5, 100)
y = stats.norm.cdf(x, 0, 1)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('CDF of X')
plt.show()
```
